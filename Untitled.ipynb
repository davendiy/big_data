```python
import re

BAD_PAT = re.compile(r'[^A-Za-zА-Яа-я\' і.]')

with open('hiba_revut.txt', 'r') as file:
    text = file.read()

nice_text = BAD_PAT.sub(' ', text)
words = nice_text.lower().split()
sentences = nice_text.lower().split('.')
```

```python
def to_one_hot(words, dictionary):
    n = len(dictionary)
    m = len(words)
    res_arr = np.zeros((m, n))
    for i, el in enumerate(words):
        res_arr[i, dictionary[el]] = 1
    return res_arr


dictionary = {el: i for i, el in enumerate(set(words))}
inv_dict = {i: el for el, i in dictionary.items()}

res_arr = to_one_hot(words, dictionary)
print(words[0], inv_dict[np.argmax(res_arr[0])])       # should be the same
```

```python
def to_bag_of_words(sentence, dictionary):
    sens_words = sentence.split()
    arr = to_one_hot(sens_words, dictionary)
    res_arr = np.sum(arr, axis=0)
    return res_arr

bag_of_words = [to_bag_of_words(sentence, dictionary) for sentence in sentences if sentence]
bag_of_words = np.array(bag_of_words)
```
